Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Datasets

```
sns.get_dataset_names()
```

Loat Datasets

```
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
tips
```

Scatter Plot

```
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'day', size = 'size')
```

Histogram Plot

```
sns.histplot(data = tips['tip'], kde = True, bins = 10)
sns.displot(tips['tip'], kde = True, bins =50)
```

Barplot

```
sns.barplot(x = 'sex', y = 'tip', data = tips, palette='YlGnBu')
```

```
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette='YlGnBu')
```

Joint plot

```
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, kind = 'kde', shade = True)
# reg/hex/kde
```

Pair plot

```
  sns.pairplot(tips, hue = 'sex')
```

Heatmap

```
num_titanic = titanic.select_dtypes(include = np.number)
sns.heatmap(num_titanic.corr(), annot = True, cmap = 'YlGnBu')
```

```
num_tips = tips.select_dtypes(include = np.number)
sns.heatmap(num_tips.corr(), annot = True, cmap = 'YlGnBu')
```

clustermap

```
sns.clustermap(num_tips)
```
