```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a Vector

```python
plt.quiver(0,0,4,8)
plt.show()
```

```python
plt.quiver(0,0,4,8,angles='xy',scale_units='xy',scale=1, color='r')
plt.quiver(0,0,2,2,angles='xy',scale_units='xy',scale=1, color = 'b')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Adding two vectors

```python
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,-2,1])
sum = v1+v2
print(sum)
```

```python
plt.quiver(0,0,v1[2],v1[3],angles='xy',scale_units='xy',scale=1, color='r')
plt.quiver(0,0,v2[2],v2[3],angles='xy',scale_units='xy',scale=1, color = 'b')
plt.quiver(0,0,sum[2],sum[3],angles='xy',scale_units='xy',scale=1, color = 'g')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
#same with subtraction
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,-2,1])
sum = v1-v2
print(sum)
```

```python
plt.quiver(0,0,v1[2],v1[3],angles='xy',scale_units='xy',scale=1, color='r')
plt.quiver(0,0,v2[2],v2[3],angles='xy',scale_units='xy',scale=1, color = 'b')
plt.quiver(0,0,sum[2],sum[3],angles='xy',scale_units='xy',scale=1, color = 'g')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

Scalar Product

```python
v1 = np.asarray([2,3])
v2 = 2*v1
print(v2)
```

```python
plt.quiver(0,0,v2[0],v2[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.show()
```

Product of two Vectors

```python
#dot product
v1 = np.asarray([0,0,2,3])
v2 = np.asarray([0,0,-2,1])
product = np.dot(v1,v2)
print(product)
```

```python
#cross product
v1 = np.asarray([2,3,4])
v2 = np.asarray([1,5,3])
product = np.cross(v1,v2)
print(product)
```

projection of Vectors

```python
mag = np.sqrt(np.sum(v2**2))
projectv1 = np.dot(v2,v1)/mag*v2
print(projectv1)
```

Matrix Addition and Subtraction

```python
m1 = np.random.randint(1,10,(3,3))
m2 = np.random.randint(1,10,(3,3))
print(m1)
print(m2)
```

```python
summ = m1+m2
subm = m1-m2
print(summ)
print(subm)
```

```python
m1 = np.random.randint(1,20, (3,5))
m2 = np.random.randint(1,20, (1,3))
print(m1)
print(m2)
```

```python
prod = np.dot(m2,m1)
print(prod)
```

```python
prod = np.multiply(4,m2)
print(prod)
```
